```python
from Bio.Seq import Seq
import matplotlib.pyplot as plt

# Function to generate k-mers from a DNA sequence
def generate_kmers(sequence, k):
    kmers = []
    for i in range(len(sequence) - k + 1):
        kmer = sequence[i:i+k]
        kmers.append(kmer)
    return kmers

# Example DNA sequence
dna_sequence = "ATCGATCGATCGATCG"

# Generate k-mers of length k=3
k = 3
kmers = generate_kmers(dna_sequence, k)

# Count frequencies of each k-mer
kmer_counts = {}
for kmer in kmers:
    if kmer in kmer_counts:
        kmer_counts[kmer] += 1
    else:
        kmer_counts[kmer] = 1

# Plotting the k-mer frequencies
plt.figure(figsize=(10, 6))
plt.bar(kmer_counts.keys(), kmer_counts.values(), color='skyblue')
plt.title(f'K-mer frequency (k={k})')
plt.xlabel('K-mer')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
